```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('/content/Screen Shot 2023-08-10 at 19.59.55.png',0)
plt.imshow(img,'gray')
```

```python
# Âm bản
negation = 255 - img
plt.imshow(negation, cmap='gray')
```

```python
# Log
logimg = np.log(img)
plt.imshow(logimg, cmap='gray')
```

```python
ivLog = 1/np.log(img)
plt.imshow(ivLog, cmap='gray')
```
